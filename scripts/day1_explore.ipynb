```python
import pandas as pd

print("=" * 50)
print("DAY 1 — Exploring IPL Data with Pandas")
print("=" * 50)

# Load data
df = pd.read_csv("/home/jovyan/matches.csv")
```

```python
# Basic exploration
print(f"\n Total matches in dataset: {len(df)}")
print(f" Columns: {list(df.columns)}")
print(f" Date range: {df['date'].min()} to {df['date'].max()}")
```

```python
df
```

```python
print("\n--- First 5 rows ---")
print(df.head())
```

```python
print("\n--- Null values per column ---")
print(df.isnull().sum())
```

```python
print("\n--- Top 5 winning teams ---")
top_winners = df['winner'].value_counts().head(5)
print(top_winners)
```

```python
print("\n--- Toss decision breakdown ---")
print(df['toss_decision'].value_counts())
```

```python
print("\n--- SRH Matches breakdown ---")
srh_matches =  df[(df['team1'] == 'Sunrisers Hyderabad') | (df['team2'] == 'Sunrisers Hyderabad')].shape[0]
print(f"SRH played {srh_matches} matches")

# Filter for 2018 matches
srh_matches_2018 = df[(df['season'] == '2018') & ((df['team1'] == 'Sunrisers Hyderabad') | (df['team2'] == 'Sunrisers Hyderabad'))].shape[0]
print(f"SRH played {srh_matches_2018} matches in 2018")

#Filter for 2018 matches where SRH won
srh_wins_2018= df[(df['season'] == '2018') & (df['winner'] == 'Sunrisers Hyderabad')].shape[0]
print(f"SRH won {srh_wins_2018} matches in 2018")
```

```python
#Ipl points table for the 2024 year
points_table_2024 = df[df['season'] == '2024'].groupby('winner').size().reset_index(name='wins').sort_values(by='wins', ascending=True)
print("\n--- IPL Points Table 2024 ---")
print(points_table_2024)
```
